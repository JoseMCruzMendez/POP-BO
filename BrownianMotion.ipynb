```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class BrownianMotion:
    def __init__(self, scale=0.1, prob=0.5):
        self.scale = scale
        self.step_size = np.sqrt(scale)
        self.prob = prob
    def step(self):
        value = np.random.uniform(0,1)
        increment = 1 if value <= self.prob else -1
        return increment
    
    def simulate(self, t):
        num_steps = int(t / self.scale)
        values = []
        for _ in range(num_steps):
            values.append(self.step())
        values = np.array(values)
        return np.cumsum(values) * self.step_size
```

```python
fig, axs = plt.subplots(3,3, figsize=(15,15))
axs = axs.ravel()
scale = 0.001
max_t = 100
T = np.linspace(0, max_t, int(max_t/scale))
for ax in axs:
    simulator = BrownianMotion(scale=scale, prob=0.499)
    values = simulator.simulate(max_t)
    ax.plot(T, values)
    simulator = BrownianMotion(scale=scale, prob=0.50)
    values = simulator.simulate(max_t)
    ax.plot(T, values)
```
